## Canny Edge detector

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../dataset/img/full_dog.jpeg')
plt.imshow(img)
plt.show()
```

```python
# Canny Edge Detector
# used middle value of 0 - 255 for both threshold
edges = cv2.Canny(image = img, threshold1 = 127, threshold2 = 127)
plt.imshow(edges)
plt.show()
```

```python
# Formula for setting thresholds for canny detector
middle_value = np.median(img) 
middle_value
```

```python
# Lower threshold - Maximum of 0 or 70 % of the middle value of the image
th1 = int(max(0, 0.7 * middle_value))
th1
```

```python
# Upper threshold - Minimum of 255 or 130 % of the middle value of the image
th2 = int(min( 255, 1.3 * middle_value))
th2
```

```python
edges = cv2.Canny(img, th1, th2)
plt.imshow(edges)
plt.show()
```

```python
# Blurring image will provide us almost perfect edge detection pattern in the image

blurred = cv2.blur(img, ksize = (3,3))
```

```python
edges = cv2.Canny(blurred, th1, th2)
plt.imshow(edges)
plt.show()
```
