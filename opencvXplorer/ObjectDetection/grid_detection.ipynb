## Grid Detection

### Chess Grid Pattern

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../dataset/img/flat_chessboard.jpg')
plt.imshow(flat_chess)
plt.show()
```

```python
# opencv has a method to find Chessboard corners only applicable for images with proper chessboard grid not fpr partial chessboard images.

# 1 - image
# 2 - grid size we need to validate/analyze the pattern
found, corners = cv2.findChessboardCorners(flat_chess, (7,7))
found # whether the image passes the criteria or not
```

```python
# Now, we are going to draw chessboard corners
flat_chess = cv2.drawChessboardCorners(flat_chess, (7,7), corners, found)
plt.imshow(flat_chess)
plt.show()
```

### Dot Grid Pattern

```python
dots = cv2.imread('../dataset/img/dot_grid.jpg')
plt.imshow(dots)
plt.show()
```

```python
# opencv has a method to find dot pattern grid corners only applicable for images with proper dot grid not fpr partial dot images.

# 1 - image
# 2 - grid size we need to validate/analyze the pattern
# 3 - kind of dot
found, corners = cv2.findCirclesGrid(dots, (6,6), cv2.CALIB_CB_SYMMETRIC_GRID)
found
```

```python
# draw pattern on the dot image
dots = cv2.drawChessboardCorners(dots, (6,6), corners, found)
plt.imshow(dots)
plt.show()
```
