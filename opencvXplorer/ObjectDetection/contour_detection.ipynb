## Contour Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../dataset/img/internal_external.jpg',0) # Gray scale reading
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# Image has grayscale variations, so applying a binary threshold ensures contours are clearly defined.
_, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
# we are going to find internal and external contours in the image 

# 1 - image
# 2 - which contour method (Here, Complete contour graps both external and internal)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
print("Type and length of the contour : ", type(contours), ", ", len(contours))
print("Type of the hierarchy : ", type(hierarchy))
```

```python
# Drawing external contours using contours and hierarchy
# Empty Array (blackboard)
external_contours = np.zeros(img.shape)
external_contours.shape
```

```python
# drawing external contours in the backboard image

for i in range(len(contours)):
    
    # Finding External Contour 
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours, contours, i, 255, 2)
```

```python
plt.imshow(external_contours, cmap='gray')
plt.show()
```

```python
# Drawing Internal contours

# step 1 - black img to draw contours
internal_contours = np.zeros(img.shape)

# step 2 - draw contours
for i in range(len(contours)):

    # Finding internal contours
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_contours, contours, i, 255, -1)

plt.imshow(internal_contours, cmap='gray')
plt.show()
```
