## Blurring and Smoothing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load a image through function
def load_img():
    img = cv2.imread('../dataset/img/bricks.jpg').astype(np.float32)/255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
# display a image through function
def show(img):
    fig = plt.figure(figsize=(12,10)).add_subplot(111)
    fig.imshow(img)
    plt.show()
```

### Gamma Correction

```python
i = load_img()
# show(i)
```

```python
# Gamma Correction Method as the gamma value increase the image gets brighter
gamma = 2

result = np.power(i, gamma)
show(result)
```

### Kernel lowpass filter (Manually)

```python
# Blurring
img = load_img()
# adding text in the image that makes easier for us to differentiate the changes
img = cv2.putText(img, "DEMON", (100,1000), cv2.FONT_HERSHEY_COMPLEX, 15, (255,0,0), 8)
# show(img)
```

```python
# setting up a Kernel method for lowpass filter
# To be compatible with image float is used for the filter kernal
kernal = np.ones((5,5), np.float32)/25
```

```python
# adding 2D filter
# second paramater - desire depth - what kind of image depth you want as output
# Example: -1 means the input depth is being assigned to the output depth)

# we can play around with Kernal size to get different results
destination = cv2.filter2D(img, -1, kernal)
show(destination)
```
