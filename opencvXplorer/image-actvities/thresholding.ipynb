## Thresholding

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading an image with gray scale
img = cv2.imread('../dataset/img/ranibow.jpg',0) # another way: cv2.IMREAD_GRAYSCALE
plt.imshow(img, cmap='gray')
plt.show()
```

```python
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV) 
```

```python
plt.imshow(thresh1, cmap='gray')
plt.show()
```

```python
img = cv2.imread('../dataset/img/newspaper.jpg',0)
plt.imshow(img)
plt.show()
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15)).add_subplot(111)
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 140, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
# Image is not clearly shows the exact letters because it was all affected by the threshold bar.
# we can use Adaptive threshold method to achieve a better image
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(th1, 0.6, th2, 0.4, 0)
show_pic(blended)
```
