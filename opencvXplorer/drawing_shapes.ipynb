## Drawing on Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Blank image is created using zeros method with the width and height of 512 and 3 channels.
# Simply a blackboard with shape of (512,512,3) --> remeber 0 - No color
blank_img = np.zeros((512,512,3), np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
plt.show()
```

#### Drawing a Rectangle over the blank image,
we have method named rectangle in cv2 lib
which takes image(where we going to draw), pt1 (Top left corner), pt2(Bottom right corner), color, thickness.

##### Note: once we draw something on the image we cannot revert it back either we create to use new blank image or reset the image.

```python
cv2.rectangle(blank_img, (300,50), (400,250), (0,255,0), 10)
plt.imshow(blank_img)
plt.show()
```

### Drawing a Square


```python
cv2.rectangle(blank_img, (200,200), (100,300), (0,255,0), 10)
plt.imshow(blank_img)
plt.show()
```

### Drawing a Circle


we have method named rectangle in cv2 lib
which takes image(where we going to draw), centre, radius, colour, thickness

```python
cv2.circle(blank_img, (400, 400), 50, (0,0,255), 10)
plt.imshow(blank_img)
plt.show()
```

### Filling Colors to the shapes

```python
# we need to provide thickness as -1
cv2.circle(blank_img, (100,80), 50, (255, 0 ,0) , -1)
plt.imshow(blank_img)
plt.show()
```

### Important Note:

As we can already seen that how we can provide shapes different color because we initalize the blank image as 3 dimensional/channels.
So in the color parameter we are something like (R,G,B) -> (0,0,255) outcome for this is Blue.
0 --> No color
255 --> full color

### Draw a line


```python
# similar as rectangle
cv2.line(blank_img, (250,250), (250,0), (102,255,255), 5)
plt.imshow(blank_img)
plt.show()
```

### Adding a Text

PutText takes image, text (what we want to show), org - placement of the text, font type, font size, color, thickness, line type

```python
cv2.putText(blank_img, "Zoro", (10, 500), cv2.FONT_HERSHEY_SIMPLEX, 4, (255,255,255), 3, cv2.LINE_AA)
plt.imshow(blank_img)
plt.show()
```

### Drawing Custom Polygon

```python
# Now we have lot in the blank img, so reseting the image
blank_img = np.zeros((512,512, 3), np.int32)
blank_img.shape
```

```python
# to draw a polygon, we need to decide the vertices(x,y coordinates of each points/edges) of the polygon
vertices = np.array([ [100,350], [200,200], [400,300], [250,450]], np.int32)
```

```python
vertices.shape
```

```python
# It's in two dimensional and no color channels available, so we need to reshape the vertices to have color channels
points = vertices.reshape((-1,1,2))
points.shape
# Acutallly, its keeping the same information but just adding a new dimension for color channel. 
```

#### Polylines -> params(image, list of vertices, isclosed bool, color, thickness)

```python
cv2.polylines(blank_img, [points], True, (255,0,0), 5)
plt.imshow(blank_img)
plt.show()
```
