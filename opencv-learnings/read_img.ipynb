## Read image with OpenCv and show through Matplotlib




Open source Computer Vision is a library mainly aimed at real-time computer vision. Created by Intel in 1999, it's written in C++.

#### Goals:
1. Able to open image with openCV.
2. Draw a simple geometrics on images.
3. Directly interact with image through callbacks.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

#### CV2 imread()

Using cv2.imread() method we are able to load the image.
It will return numpy array of the provided image.
This reduce the use the PIL library to open image and convert to numpy array. 

Note: this method doesn't return any error if we provide unknown/wrong path instead it will return NoneType.

```python
img = cv2.imread("../img/vicky.jpg")
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

#### Important Note

##### Matplotlib --> Read image as RED GREEN BLUE channels.
##### OpenCV     --> Read image as BLUE GREEN RED channels

That's why the above result looks like red replaces blue and blue replaces red.

```python
# converting Blue Green Red image to Red Green Blue
# because plt read image as Red Greeen Blue

fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
plt.show()
```

### Adding grey scale to the image while reading

```python
img_gray = cv2.imread("../img/vicky.jpg", cv2.IMREAD_GRAYSCALE)
img_gray.shape

# It is no longer a three dimensional image because when using IMREAD_GRAYSCALE , the image not be decoded as *B G R* channels.
```

```python
# The array values ranges from 0(no colour) - 255(full color)

print("Min : ",img_gray.min())
print("Max : ",img_gray.max())
```

```python
plt.imshow(img_gray) # It shows the image as no colour area in the image as bluish and light color area in the image as yellowish green.
plt.show()
```

```python
# to suit the grey scale, we need to use cmap = 'grey' 
plt.imshow(img_gray, cmap='gray')
plt.show()
```

### Resizing Image

```python
fix_img.shape
```

```python
resized_img= cv2.resize(fix_img, (500,500))
plt.imshow(resized_img)
plt.show()
```

```python
# Resizing with ratio
# now we are reducing 50% of the image 
w_ratio = 0.5
h_ratio = 0.5

resized_img_with_ratio = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
plt.imshow(resized_img_with_ratio)
plt.show()
```

### Fliping Image


```python
# flipping horizontal

h_flip = cv2.flip(fix_img, 0)
plt.imshow(h_flip)
plt.show()
```

```python
# Flipping Vertical

h_flip = cv2.flip(fix_img, 1)
plt.imshow(h_flip)
plt.show()
```

```python
# Flipping both sides at once

h_flip = cv2.flip(fix_img, -1)
plt.imshow(h_flip)
plt.show()
```

### Saving Image file

```python
cv2.imwrite('totally_new.jpg', fix_img)
```
