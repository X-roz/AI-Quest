## BITWISE OPERATIONS

1. OR Gate
2. AND Gate
3. NOT Gate
4. XOR Gate

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Black image with white rectangle
b = np.zeros((300,300, 3), np.uint8)
b = cv2.rectangle(b, (0,0), (200,50), (255,255,255), -1 )
plt.imshow(b)
plt.show()
```

```python
# White image with black rectangle
w = np.full((300,300, 3), 255, np.uint8)
w = cv2.rectangle(w, (0,0), (200,50), (0,255,0), -1)
plt.imshow(w)
plt.show()
```

### Bitwise AND Operation

```python
# And Operation on both images
and_gate = cv2.bitwise_and(b, w)
plt.imshow(and_gate)
plt.show()
```

### Bitwise OR Operation

```python
# to get more understanding add a black rectangle somewhere in the img2
w = cv2.rectangle(w, (150,250), (300,300), (0,0,0), -1)
```

```python
# OR Operation on images
or_gate = cv2.bitwise_or(w, b)
plt.imshow(or_gate)
plt.show()
```

### Bitwise NOT Operation

```python
# NOT Operation
not_gate = cv2.bitwise_not(b)
plt.imshow(not_gate)
plt.show()
```

### Bitwise XOR Operation

```python
# XOR Operation
xor_gate = cv2.bitwise_xor(b,w)
plt.imshow(xor_gate)
plt.show()
```
